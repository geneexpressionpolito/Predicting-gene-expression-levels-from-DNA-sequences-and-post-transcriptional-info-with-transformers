```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_trainmicro = scaler.fit_transform(tabellone_train)
X_valmicro = scaler.transform(tabellone_val)
X_testmicro = scaler.transform(tabellone_test)
```

```python
from sklearn.decomposition import TruncatedSVD
svd = TruncatedSVD(n_components=1000, random_state=42)
X_trainmicro = svd.fit_transform(X_trainmicro)
X_valmicro = svd.transform(X_valmicro)
X_testmicro = svd.transform(X_testmicro)
print(svd.explained_variance_ratio_)

print(svd.explained_variance_ratio_.sum())

print(svd.singular_values_)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
clf = RandomForestRegressor(random_state=50)
clf.fit(X_trainmicro, tuttiLabel_train)
# y_pred = clf.predict(X_val))
# print(classification_report(y_pred, y_val))
y_pred = clf.predict(X_testmicro)
print(r2_score(tuttiLabel_test, y_pred))
```

```python
import tensorflow as tf

inputs = tf.keras.layers.Input(shape=(1000,))
x = tf.keras.layers.Dense(64, activation="relu")(inputs)
x = tf.keras.layers.Dropout(0.00099)(x)
x = tf.keras.layers.Dense(2, activation="relu")(x)
x = tf.keras.layers.Dropout(0.01546)(x)
predictions = tf.keras.layers.Dense(1, activation = 'linear')(x)
model = tf.keras.Model(inputs=inputs, outputs=predictions)
optimizer = tf.keras.optimizers.SGD(lr=5e-4, momentum=0.9)
model.compile(optimizer=optimizer, loss='mean_squared_error')

model.fit(X_trainmicro, tuttiLabel_train, epochs=50, validation_data = (X_valmicro, tuttiLabel_val), batch_size=256, shuffle=True)
```

```python
from scipy import stats
predictions = model.predict(X_testmicro).flatten()
slope, intercept, r_value, p_value, std_err = stats.linregress(predictions, tuttiLabel_test)
print('Test R^2 = %.3f' % r_value**2)
```
